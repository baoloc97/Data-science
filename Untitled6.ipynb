```python
import easyocr
import cv2
from matplotlib import pyplot as pyplot
import numpy as np
```

```python
img = cv2.imread('image/3.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_new = cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
width = 900
height = 900 # keep original height
dim = (width, height)
resized = cv2.resize(gray_new, dim, interpolation = cv2.INTER_AREA)
pyplot.imshow(resized)
```

```python


reader = easyocr.Reader(['en'],gpu=True)
result = reader.readtext(resized)
```

```python
text= [result[i][1] for i in range(len(result))]
print(text)
```
